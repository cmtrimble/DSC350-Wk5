Description: This program ingests CSVs for various company stocks. We add a column, append the files, and print to new file. We also change the datatypes for two columns, find the smallest values, and melt columns together.  
Codes in this program have been adapted from Hands-On Data Analysis with Pandas - Second Edition (Molin S., 2021)

```python
# Problem 1
import pandas as pd

# Reading the CSV files
df = pd.read_csv('aapl.csv')
df1 = pd.read_csv('amzn.csv')
df2 = pd.read_csv('fb.csv')
df3 = pd.read_csv('goog.csv')
df4 = pd.read_csv('nflx.csv')

# List of DataFrames and their corresponding filenames
dataframes = [df, df1, df2, df3, df4]
filenames = ['aapl.csv', 'amzn.csv', 'fb.csv', 'goog.csv', 'nflx.csv']

# Searching for filename in the DataFrames.
for df, filename in zip(dataframes, filenames):
    ticker = filename.replace('.csv', '').upper()  # Staging the values that are supposed to go in ticker column. Removing the file extension from the filename for visual appeal.
    df['ticker'] = ticker  # Adds column to DF. 

combined_df = pd.concat(dataframes, ignore_index=True)  # Merges the DataFrames.
combined_df.to_csv('faang.csv', index=False)  # Prints the new file 'faang' to a csv file.
```

```python
# Problem 2
import pandas as pd

df = pd.read_csv('faang.csv')

# Defined the column we want to modify, defines the format for the datetime.
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')  
df['volume'] = df['volume'].astype(int)  # Sets the datatype for volume as an integer. 
df.sort_values(by=['date', 'ticker'], ascending=[False, True]).head(10)  # Sorts values by the defined values and sets the direction they're sorted.

print(df)
```

```python
# Problem 3
# Finds the rows with the lowest volume values, and sets the number of rows we're looking for.
df.nsmallest(n=7, columns=['volume'])
```

```python
# Problem 4
# Defines the ID variables and the variables we want to "melt" into one column.
df.melt(id_vars=['date', 'ticker'], value_vars=['open', 'high', 'low', 'close', 'volume'])
```
